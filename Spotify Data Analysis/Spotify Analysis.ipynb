```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks=pd.read_csv("tracks.csv")
df_tracks
```

```python
df_tracks.head() #to fetech first 5 rows we .head()
```

```python
##check null values
pd.isnull(df_tracks).sum()
```

```python
# to check columns and datatypes 
df_tracks.info()
```

```python
#to find 10 least popular songs
sorted_df=df_tracks.sort_values('popularity',ascending=True).head(10)
sorted_df
```

```python
#quick overview of the statistical properties of a dataset. 
df_tracks.describe().transpose()
```

```python
#top 10 songs popularity greater than 90
popular = df_tracks.query("popularity>90",inplace=False).sort_values('popularity', ascending=False)
popular[1:11]

#we can use head or indexing for top songs
#inplace=false if we want new data frame rther than updating the older one
#.query is used for the creteria passing
```

```python
#to set new index

df_tracks.set_index("release_date" , inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# to check artists in the some specific loaction
df_tracks[["artists"]].iloc[9]
#df_tracks.loc[5,"artists"]
```

```python
# convert the durration into sec by dividing 1000
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1) # axis =1 is for row and aaxis = 0 we use for column
df_tracks.duration.head()   
```

```python
corr=df_tracks.corr()
corr
```

```python
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")

plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks))) 
len(sample_df)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='loudness', x='energy', color='c').set(title="Loudness vs Energy Correaltion")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='popularity', x='acousticness', color='b').set(title="popularity vs acousticness Correaltion")
```

```python
 df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks['dates'] = pd.to_datetime(df_tracks['dates'])
years = df_tracks['dates'].dt.year
```

```python
total_dr =  df_tracks.duration
fig,ax=plt.subplots(figsize=(18,7))
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False)
plt.title("Year vs Duration")
plt.xticks(rotation=90)
plt.show()

#ax=ax to place error bars on the correct loaction
#errwidth=False to hide the error bars
```

```python
total_dr=df_tracks['duration']
sns.set_style(style="whitegrid")
fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(x=years, y=total_dr, ax=ax)
plt.title("Year vs Duration")
plt.xticks(rotation=90)
plt.show()
```
